Смотрим кратко на данные нашего датасета

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = '../data/_data.csv'
df = pd.read_csv(path)
```

```python
df.head() #Верхние значения датасета
```

```python
df.sample(5) # Смотрим на случайный срез нашего датасета
```

```python
# Смотрим краткое техническое описание нашего датасета
df.info()
```

```python
# Краткая статистика по нашему датасету
df.describe()
```

```python
df.describe(include='object')
```

```python
missings = df.isna().sum()
totals = df.shape[0]

fig, ax = plt.subplots(figsize=(10,6))

ax.bar(missings.index, (totals - missings)/totals, label='Заполненное')
ax.bar(missings.index, missings/totals,bottom=(totals - missings)/totals, label='Пустое')

ax.set_title('Соотношение пустых и заполненных значений в каждом столбце датафрейма')
ax.set_xlabel('Столбец')
ax.set_ylabel('Пропорция')

ax.set_ylim(0, 1.1)

ax.set_xticks(missings.index)
ax.set_xticklabels(missings.index, rotation=45, ha='right')

ax.legend(title='Значение')

plt.tight_layout()
plt.savefig('missings.png')
plt.show()
```

```python
uniques = df.nunique()
totals = df.shape[0]

fig, ax = plt.subplots(figsize=(10,6))

ax.bar(uniques.index, (totals - uniques) / totals, label='Дублированное')
ax.bar(uniques.index, uniques / totals, bottom=(totals - uniques) / totals, label='Уникальное')

ax.set_title('Соотношение уникальных и дублированных значений в каждом столбце датафрейма')
ax.set_xlabel('Столбец')
ax.set_ylabel('Пропорция')

ax.set_ylim(0, 1.1)

ax.set_xticks(uniques.index)
ax.set_xticklabels(uniques.index, rotation=45, ha='right')

ax.legend(title='Значение', loc='upper right')

plt.tight_layout()
plt.savefig('uniques.png')
plt.show()
```
